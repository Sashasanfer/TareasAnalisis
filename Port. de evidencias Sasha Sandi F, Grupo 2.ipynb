<h3 Style= "color:#20dad8"> 4. Manejo de valores Faltantes </h3>

<p Style= 'text-align: justify'> Que se hace para el manejor de valires faltantes </p>

<ol>
    <li>Identificar como se muestran los valores faltantes</li>
    <li>Contar la cantidad de valores faltantes en el set de datos</li>
    <li>Imputacion de valores numericos</li>
    <li>Imputacion de valores categoricos</li>
    <li>Eliminar filas </li>
<ol>   


<h5 Style= "color:#20dad8"> 4.1 Identificar valores Faltantes </h5>

```python
dfAutos.replace("?", np.nan, inplace = True)
dfAutos.head(5)
```

<h5 Style= "color:#20dad8"> 4.2 Contar la cantidad de valores Faltantes </h5>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h5 Style = "color:#20dad8"> 4.3 Imputacion de valores numericos </h5>

```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0) #busco la media
print("La media es:", av_normalized)
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
print("La media es:", av_bore)
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
print("La media es:", av_stroke)
av_horsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
print("La media es:", av_horsepower)
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
print("La media es:", av_peak)
```

```python
dfAutos["normalized"].replace(np.nan, 122, inplace = True)
dfAutos["bore"].replace(np.nan, 3.31, inplace = True)
dfAutos["stroke"].replace(np.nan, 3.27, inplace = True)
dfAutos["horsepower"].replace(np.nan, 102, inplace = True)
dfAutos["peak"].replace(np.nan, 5100, inplace = True)
dfAutos.head(5)
```

<h5 Style = "color:#20dad8"> 4.4 Imputacion de valores categoricos </h5>

```python
dfAutos["doors"].value_counts().idxmax() #Busco el valor mas frecuente
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace = True)
```

<h5 Style = "color:#20dad8"> 4.5 Eliminacion de filas con valores faltantes </h5>

```python
dfAutos.dropna(subset = ["price"], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
```

```python
dfTemp = dfAutos.isnull()
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h3 Style= "color:#20dad8"> 5. Estandarizacion de los datos </h3>

<ol>
    <li>Verificar en Dominio de las variables</li>
    <li>Convertir valores númericos, en funcion de la interpretacion del medio</li>
    <li>Normalizar valores entre 0 y 1</li>
    <li>Creación de rangos de valores (binning)</li>
<ol>


<h5 Style= "color:#20dad8"> 5.1 Verificar el Dominio de las Variables </h5>

```python
["normalized"],["bore"],["stroke"],["horsepower"],["peak"]
```

```python
dfAutos[["normalized","horsepower","peak","price"]] = dfAutos[["normalized","horsepower","peak","price"]].astype("int")
dfAutos[["bore","stroke"]] = dfAutos[["bore","stroke"]].astype("float")
dfAutos.dtypes
```

```python
dfAutos.describe()
```

<h5 Style= "color:#20dad8"> 5.2 Convertir valores númericos, en funcion de la interpretacion del medio (Estandarizar)</h5>

```python
dfAutos.head(15)
```

```python
dfAutos["City L/100km (Calc)"] = 235/dfAutos["City"]
dfAutos["Higway L/100km (Calc)"] = 235/dfAutos["Higway"]
dfAutos.head(5)
```

<h5 Style= "color:#20dad8"> 5.3 Normalizar valores entre 0 y 1 (Lo "Incomparable") </h5>

```python
dfTemp = dfAutos[["City L/100km (Calc)", "Higway L/100km (Calc)", "engineSize"]]
dfTemp.head(10)
```

```python
dfTemp.describe()
```

```python
dfAutos["City L/100km (norm)"] = dfAutos["City L/100km (Calc)"]/dfAutos["City L/100km (Calc)"].max()

dfAutos["Higway L/100km (norm)"] = dfAutos["Higway L/100km (Calc)"]/dfAutos["Higway L/100km (Calc)"].max()

dfAutos["engineSize (norm)"] = dfAutos["engineSize"]/dfAutos["engineSize"].max()

dfTemp = dfAutos[["City L/100km (Calc)","Higway L/100km (Calc)","engineSize","City L/100km (norm)","Higway L/100km (norm)","engineSize (norm)"]]
dfTemp.head(10)
```

<h5 Style= "color:#20dad8"> 5.4 Crear rangos para agrupar datos</h5>

```python
dfTemp = dfAutos["horsepower"] 
dfTemp.describe()
```

```python
#bins = np.linspace(min(dfAutos["horsepower"]), max(dfAutos["horsepower"]), 4)
bins  = np.linspace (48,262,4)
bins
```

```python
categorias = ["Bajo","Medio","Alto",]
dfAutos["horsepower (cat)"] = pd.cut(dfAutos["horsepower"], bins, labels = categorias, include_lowest=True)
dfTemp = dfAutos[["horsepower","horsepower (cat)"]]
dfTemp.head(10)
```

Pregunta 1: ¿Cuatos Autos existen según marca?

```python
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending = False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,6.9), dpi=100)

 

eje_x = gf.index.tolist()
data = gf['symboling'].tolist()

 

bar_colors = ['#89abb4','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#fdd1b6']
ax.barh(eje_x, data, color=bar_colors)

 

#--------------------------------------------------------------------------------------------------------------------#
#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO
title = "Grafico 1: Vehiculos"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Liberation Sans Narrow")

 

#FORMATO DEL SUBTITULO
subTitle = "Cantidad de vehiculos agrupados segun fabricante"
ax.set_title(subTitle,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

 

#FORMATO DE LOS EJES
etiqueta = "Cantidad de vehiculos (ABS)"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

 

#CONFIGURACION DE COORDENADAS DE LA ETIQUETA DEL EJE X
ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)

 

#CONFIGURACION EL TAMANO DEL TEXTO DEL EJE Y
ax.tick_params(axis='both',
               labelsize=10)

 

#CONFIGURACION LA UBICACION DEL GRAFICO
plt.subplots_adjust(top=0.80)
plt.show()
```
