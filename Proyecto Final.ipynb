<h5 Style= "color:#714700">4.¿Cuál es el costo promedio de las propiedades residenciales y comerciales? </h3>

```python
residentialPro = df[df['propertyType'] == 'Residential']
commercialPro = df[df['propertyType'] == 'Commercial']

averageRes_cost = residentialPro['saleAmount'].mean()
averageComm_cost = commercialPro['saleAmount'].mean()

print(f"Costo promedio de propiedades residenciales: ${averageRes_cost:.2f}")
print(f"Costo promedio de propiedades comerciales: ${averageComm_cost:.2f}")
```

```python
categories = ['Residenciales', 'Comerciales']
average_costs = [averageRes_cost, averageComm_cost]

plt.figure(figsize=(10, 6))
plt.pie(average_costs, labels=categories, colors=['#2bc894', '#d5d467'], autopct=lambda p: f'${sum(average_costs) * p / 100:,.2f}', startangle=140, shadow=True)

plt.title('Distribución del Costo Promedio de Propiedades Residenciales y Comerciales', fontsize=16, fontweight="bold", y=1, pad=0)

plt.setp(plt.gca().get_xticklabels(), fontsize=13, fontweight="bold")
plt.setp(plt.gca().get_yticklabels(), fontsize=13, fontweight="bold")

plt.tight_layout()
plt.show()
```

<h5 Style= "color:#714700">5.¿Cuál es tipo de propiedades residenciales que se vende con mayor frecuencia, donde y con que costo promedio? </h3>

```python
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

residential_properties = df[df['propertyType'] == 'Residential']

most_common_residential_type = residential_properties['residentialType'].value_counts().idxmax()

average_cost_most_common = residential_properties[residential_properties['residentialType'] == most_common_residential_type]['saleAmount'].mean()

most_common_location = residential_properties[residential_properties['residentialType'] == most_common_residential_type]['town'].value_counts().idxmax()

print(f"El tipo de propiedad residencial más común es: {most_common_residential_type}")
print(f"Ubicación donde se vende con mayor frecuencia: {most_common_location}")
print(f"Costo promedio de las propiedades residenciales de tipo '{most_common_residential_type}': {locale.currency(average_cost_most_common, grouping=True)}")
```

<h5 Style= "color:#714700">6.¿Cuál es margen promedio de ganancia por cada una de las propiedades vendidas? </h3>

```python
#Esclui montos de venta igual a cero para evitar la division entre 0 de la formula
filtered_df = df[df['saleAmount'] > 0]

filtered_df['profit_margin'] = (filtered_df['saleAmount'] - filtered_df['assessedValue']) / filtered_df['saleAmount']

average_profit_margin = filtered_df['profit_margin'].mean()

print(f"Margen promedio de ganancia por cada propiedad vendida: {average_profit_margin:.2%}")
print("Un margen de ganancia negativo sugiere que las propiedades se vendieron con pérdidas en promedio")
```

```python
filtered_df = df[df['saleAmount'] > 0]

filtered_df.loc[:, 'profit_margin'] = (filtered_df['saleAmount'] - filtered_df['assessedValue']) / filtered_df['saleAmount']

average_profit_margin_by_type = filtered_df.groupby('propertyType')['profit_margin'].mean()

plt.figure(figsize=(12, 8))
ax = sns.barplot(x=average_profit_margin_by_type.index, y=average_profit_margin_by_type.values, palette='Blues')

plt.xlabel('Tipo de Propiedad', fontsize=13, fontweight='bold')
plt.ylabel('Margen de Ganancia Promedio', fontsize=13, fontweight='bold')
plt.title('Margen de Ganancia Promedio por Tipo de Propiedad', fontsize=16, fontweight='bold')
plt.xticks(rotation=45)

for p in ax.patches:
    x = p.get_x() + p.get_width() / 2.
    y = p.get_height()
    if p.get_height() > 0:
        y += 0.005
    ax.annotate(f'{y:.2%}', (x, y),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 14),
                textcoords='offset points', fontweight='bold', style='italic')

x_mid = sum(ax.get_xlim()) / 2
y_mid = sum(ax.get_ylim()) / 2

average_annotation = f'Promedio de Ganancia: -162.86%'
ax.annotate(average_annotation, (x_mid, y_mid),
            ha='center', va='center', fontsize=12, color='black',
            bbox=dict(boxstyle='square,pad=0.3', facecolor='lightgray', edgecolor='gray'))

plt.tight_layout()
plt.show()
```
